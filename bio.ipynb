```python
import torch
from transformers import AutoTokenizer, AutoModel, BertConfig
from sklearn.metrics.pairwise import cosine_similarity
```

```python

tokenizer = AutoTokenizer.from_pretrained("zhihan1996/DNABERT-2-117M", trust_remote_code=True)
config = BertConfig.from_pretrained("zhihan1996/DNABERT-2-117M")
model = AutoModel.from_config(config)
```

```python
dna = "ACGTAGCATCGGATCTATCTATCGACACTTGGTTATCGATCTACGAGCATCTCGTTAGC"
dna2 = "TACGCTAGATCGTAGCTAGCATCGATCGATCGTAGCTAGCATCGTAGCATCGTAGCATCGT"
inputs = tokenizer(dna, return_tensors = 'pt')["input_ids"]
inputs2 = tokenizer(dna2, return_tensors = 'pt')["input_ids"]

print(dna)
print(inputs)

for i in range(1, len(inputs[0]) - 1):
    print(tokenizer.decode(inputs[0][i]), end="-")

print()
print()
print()
print()

print(dna2)
print(inputs2)
for i in range(1, len(inputs2[0]) - 1):
    print(tokenizer.decode(inputs2[0][i]), end="-")

output1 = model(inputs)[0] # [1, sequence_length, 768]
output2 = model(inputs2)[0] # [1, sequence_length, 768]


```

```python
similarity = cosine_similarity(output1[0].detach().numpy(),output2[0].detach().numpy())
```

```python

similarity_arr = []
for i in range(1, len(inputs[0]) - 1):
    for j in range(1, len(inputs2[0]) - 1):
        similarity_arr.append(((i,tokenizer.decode(inputs[0][i])), (j,tokenizer.decode(inputs2[0][j])), similarity[i][j]))
      
print(similarity_arr)
```

```python
sorted_similarity_arr = sorted(similarity_arr, key=lambda x: x[2], reverse=True)
```

```python
for item in sorted_similarity_arr:
    print(item)
```
